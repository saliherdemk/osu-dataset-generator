# Hit Objects


```python
cols = ['x', 'y', 'repeat', 'slider_length', 'spinner_time',
       'path_1', 'path_2', 'path_3', 'path_4', 'path_5', 'path_6', 'path_7', 'path_8', 'path_9','path_10', 'path_11', 'path_12', 'path_13', 'path_14', 'path_15',
       'path_16', 'path_17', 'path_18', 'path_19', 'path_20', 'path_21', 'path_22', 'path_23', 'path_24', 'path_25', 'path_26',
       'type_circle', 'type_slider', 'type_slient', 'type_spinner',
       'hit_sound_0.0', 'hit_sound_2.0', 'hit_sound_4.0', 'hit_sound_6.0','hit_sound_8.0', 'hit_sound_10.0', 'hit_sound_12.0','hit_sound_14.0',
       'curve_type_B', 'curve_type_E','curve_type_L', 'curve_type_P', 
       'new_combo_False', 'new_combo_True',
       'time', 'beat_length']
hit_filtered = filtered[~filtered["type_slient"]][cols]
```

```python
hit_filtered
```

```python
hit_sound_cols = ['hit_sound_0.0', 'hit_sound_2.0', 'hit_sound_4.0', 'hit_sound_6.0','hit_sound_8.0', 'hit_sound_10.0', 'hit_sound_12.0','hit_sound_14.0']
hit_filtered['hit_sound'] = hit_filtered[hit_sound_cols].idxmax(axis=1).str.split("_").str[-1].astype(float).astype(int)
```

```python
curve_type_cols = ['curve_type_B', 'curve_type_E','curve_type_L', 'curve_type_P']
hit_filtered['curve_type'] = hit_filtered[curve_type_cols].idxmax(axis=1).str.split("_").str[-1]
```

```python
type_cols = ['type_circle', 'type_slider', 'type_slient', 'type_spinner']
hit_filtered['type'] = hit_filtered[type_cols].idxmax(axis=1).str.split("_").str[-1]
```

```python
combo_cols = ['new_combo_False', 'new_combo_True']
hit_filtered['new_combo'] = hit_filtered[combo_cols].idxmax(axis=1).str.split("_").str[-1]
```

```python
path_x_columns = [
    col
    for col in df.columns
    if col.startswith("path_") and int(col.split("_")[1]) % 2 != 0
]
path_y_columns = [
    col
    for col in df.columns
    if col.startswith("path_") and int(col.split("_")[1]) % 2 == 0
]
for i, (col_x, col_y) in enumerate(zip(path_x_columns, path_y_columns)):
    hit_filtered[f"point_{i}"] = (
        hit_filtered[col_x].round(0).astype(int).astype(str) + ":" + hit_filtered[col_y].round(0).astype(int).astype(str)
    )
```

```python
hit_columns = ['x', 'y', 'time', 'type', 'hit_sound', 'curve_type', "new_combo", 'repeat', 'slider_length', 'spinner_time']\
+ [col for col in hit_filtered.columns if col.startswith("point")] 
hit_filtered = hit_filtered[hit_columns]
```

```python
hit_filtered
```

```python
def rounded_str(x):
    return str(int(x))

def remove_trailing_zeros(data):
    while data and data[-1] == '0:0':
        data.pop()
    return data

def get_type_params(hit_type, new_combo, curve_type, curve_columns, repeat, slider_length, spinner_time):
    flag = 0
    objectParams = ""
    
    if hit_type == 'circle':
        flag |= (1 << 0)
    elif hit_type == 'slider':
        flag |= (1 << 1)
        curve_columns = remove_trailing_zeros(curve_columns)
        objectParams += ',' + curve_type + "|" + '|'.join(curve_columns) + ',' + rounded_str(repeat) + ',' + str(round(slider_length, 12))
    elif hit_type == 'spinner':
        flag |= (1 << 3)
        objectParams += ',' + rounded_str(spinner_time)
    
    if new_combo:
        flag |= (1 << 2)
    
    return flag, objectParams

hit_objects = []
for i, row in hit_filtered.iterrows():
    x, y, time, hit_type, hit_sound, curve_type, new_combo, repeat, slider_length, spinner_time, *curve_columns = row[hit_columns]
    flag, objectParams = get_type_params(hit_type, new_combo, curve_type, curve_columns, repeat, slider_length, spinner_time)
    hit_objects.append(rounded_str(x) + "," + rounded_str(y) + "," + rounded_str(time) \
          + ',' + str(flag) + "," + rounded_str(hit_sound) + str(objectParams)
         )
```
