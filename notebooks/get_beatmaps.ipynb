```python
from dotenv import load_dotenv
import os

load_dotenv()
API_KEY = os.getenv("API_KEY")
```

```python
import requests
def fetch_beatmaps_since(since_date, limit = 500):
    url = "https://osu.ppy.sh/api/get_beatmaps"
    params = {
        "k": API_KEY,
        "m": 0,
        "since": since_date,
        "limit": limit,
    }
    response = requests.get(url, params=params)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error: {response.status_code} - {response.text}")
        return []
```

```python
from datetime import datetime
import time

last_fetched_date = "2011-01-01 00:00:00"
max_date_formatted = datetime.strptime("2025-08-01 00:00:00", "%Y-%m-%d %H:%M:%S")

beatmapset_ids = set()

while datetime.strptime(last_fetched_date, "%Y-%m-%d %H:%M:%S") < max_date_formatted:
    beatmaps = fetch_beatmaps_since(last_fetched_date)
    
    for beatmap in beatmaps:
        beatmapset_ids.add(beatmap["beatmapset_id"])
        last_fetched_date = beatmap["approved_date"]
    time.sleep(1)
    print(last_fetched_date)
```

```python
import json
beatmapset_ids = [int(x) for x in beatmapset_ids]

with open("data.json", "w") as f:
    json.dump(list(beatmapset_ids), f)
```
