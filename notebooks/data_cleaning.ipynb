# OSU OAUTH

```python
from dotenv import load_dotenv
import requests

load_dotenv()
CLIENT_ID = os.getenv("CLIENT_ID")
CLIENT_SECRET = os.getenv("CLIENT_SECRET")


def get_access_token():
    url = "https://osu.ppy.sh/oauth/token"
    headers = {"Content-Type": "application/x-www-form-urlencoded"}
    data = {
        "client_id": CLIENT_ID,
        "client_secret": CLIENT_SECRET,
        "grant_type": "client_credentials",
        "scope": "public",
    }

    response = requests.post(url, headers=headers, data=data)
    return response.json().get("access_token")

access_token = get_access_token()
```

```python
import pandas as pd
import os

database_path = "/mnt/L-HDD/post_2010_dataset/"

beatmaps_csv = os.path.join(database_path, "beatmaps.csv")
hit_objects_csv = os.path.join(database_path, "hit_objects.csv")
time_points_csv = os.path.join(database_path, "timing_points.csv")

beatmaps_df = pd.read_csv(beatmaps_csv, parse_dates=["ranked_date"])
hit_objects_df = pd.read_csv(hit_objects_csv)
time_points_df = pd.read_csv(time_points_csv)
```

```python
empty_cells = beatmaps_df[beatmaps_df.isna().any(axis=1)]
```

```python
empty_cell_ids = set(empty_cells["beatmap_id"])
```

```python
def get_beatmapset_status(beatmapset_id, access_token):
    url = f"https://osu.ppy.sh/api/v2/beatmapsets/{beatmapset_id}"
    headers = {"Authorization": f"Bearer {access_token}", "Accept": "application/json"}

    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        beatmapset = response.json()
        beatmaps = {}
        for beatmap in beatmapset.get("beatmaps"):
            beatmaps[beatmap["version"]] = beatmap["ar"]
        return beatmaps
    else:
        return f"Error: {response.status_code} - {response.text}"
```

```python
import time
ars = {}
for id in empty_cell_ids:
    ars[id] = get_beatmapset_status(id, access_token)
    time.sleep(0.2)
```

```python
def fill_approach_rate(row):
    beatmap_id = row["beatmap_id"]
    version = row["Version"]
    
    if beatmap_id in ars:
        return ars[beatmap_id].get(version, None)
    return row["ApproachRate"]

beatmaps_df["ApproachRate"] = beatmaps_df.apply(fill_approach_rate, axis=1)
```
