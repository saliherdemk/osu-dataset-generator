```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset_path = "/mnt/L-HDD/Public/ranked/"

beatmap_df = pd.read_csv(dataset_path + "beatmaps.csv")
```

```python
beatmap_df["difficulty_rating"] = beatmap_df["difficulty_rating"].astype(int)
```

```python
difficulty_counts = beatmap_df['difficulty_rating'].value_counts().sort_index()

# Plotting
plt.figure(figsize=(8, 6))
plt.bar(difficulty_counts.index, difficulty_counts.values, color='skyblue')
plt.xlabel('Difficulty Rating')
plt.ylabel('Count')
plt.title('Count of Beatmaps by Difficulty Rating')
plt.xticks(difficulty_counts.index)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
